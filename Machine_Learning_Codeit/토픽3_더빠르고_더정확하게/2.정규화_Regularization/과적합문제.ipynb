```python
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
import pandas as pd
```

```python
df = pd.read_csv("Desktop/admission_data.csv").drop('Serial No.', axis=1)
```

```python
df
```

```python
X = df.drop(['Chance of Admit '], axis=1)
```

```python
polynomial_transformer = PolynomialFeatures(6)
```

```python
ploynomial_features = polynomial_transformer.fit_transform(X.values)
```

```python
features = polynomial_transformer.get_feature_names(X.columns)
```

```python
X = pd.DataFrame(ploynomial_features, columns=features)
```

```python
X.head()
```

```python
y = df['Chance of Admit ']
```

```python
y = pd.DataFrame(y, columns=['Chance of Admit '])
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_train_predict = model.predict(X_train)
```

```python
y_test_predict = model.predict(X_test)
```

```python
ytrain = mean_squared_error(y_train, y_train_predict) ** 0.5
```

```python
ytest = mean_squared_error(y_test, y_test_predict) ** 0.5
```
