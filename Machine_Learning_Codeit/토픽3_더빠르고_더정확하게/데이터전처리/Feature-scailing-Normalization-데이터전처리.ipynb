```python
import pandas as pd
from sklearn import preprocessing
```

```python
path = 'Desktop/NBA_player_of_the_week.csv'
```

```python
df = pd.read_csv(path)
```

```python
df.describe()
```

```python
new_df = df[['Age', 'Height CM', 'Weight KG']]
```

```python
new_df
```

```python
scaler = preprocessing.MinMaxScaler()
```

```python
normalized_data = scaler.fit_transform(new_df)
```

```python
normalized_df = pd.DataFrame(normalized_data, columns=['Age', 'height cm', 'weight kg'])
```
